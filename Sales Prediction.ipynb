```python
#importing the libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
raw_csv_data = pd.read_csv("Advertising.csv")
df = raw_csv_data.copy()
```

```python
sns.pairplot(data =df, x_vars =['TV', 'Newspaper', 'Radio'], y_vars = 'Sales', height = 5, aspect = 1, kind = 'scatter')
plt.show()
```

```python
selected_features = ['TV', 'Radio', 'Newspaper']
x = df[selected_features]
y = df.Sales
```

```python
#train test splitting
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3)
```

Prediction

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
model.predict(x_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
mode12 = DecisionTreeRegressor()
mode12.fit(x_train, y_train)
```

```python
mode12.score(x_test, y_test)
```

```python
mode12.predict(x_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
mode13 = RandomForestRegressor(n_estimators= 30)
mode13.fit(x_train, y_train)
```

```python
mode13.score(x_test, y_test)
```

```python
print("Linear Regression accuracy :", model.score(x_test, y_test)*100)
print("Decision Tree accuracy:", mode12.score(x_test, y_test)*100)
print("Random Forest accuracy:", mode13.score(x_test, y_test)*100)
```
